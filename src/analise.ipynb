```python
def verifica_desfecho(obito, piora, melhora, alta):
    if(obito > 0 ):
        return 'Óbito'
    elif(piora > 0 ):
        return 'Piora'
    elif(melhora > 0):
        return 'Melhora'
    elif(alta > 0):
        return 'Alta'
    else:
        return 'SemDesfecho'
    
df['desfecho'] = np.vectorize(verifica_desfecho)(df['obito'], df['piora'], df['melhora'], df['alta'])
df['desfecho'].value_counts()
```

```python
classe_minoritaria = df_analise.loc[df['desfecho'].isin(['Alta', 'Óbito'])]
len(classe_minoritaria)
classe_majoritaria = df_analise.loc[df['desfecho'].isin(['Melhora', 'Piora'])]
len(classe_majoritaria)
from sklearn.utils import resample
# Aplicar downsampling na classe majoritária
classe_majoritaria_downsampled = resample(classe_majoritaria,
                                          replace=False,  # Sem reposição
                                          n_samples=len(classe_minoritaria),  # Igualar o número de amostras da classe minoritária
                                          random_state=42)  # Reprodutibilidade
# Combinar as classes majoritária downsampled com a classe minoritária
df_downsampled = pd.concat([classe_majoritaria_downsampled, classe_minoritaria])
```

```python
y = df_downsampled['desfecho']
X = pd.get_dummies(df_downsampled.drop(['desfecho', 'cns_paciente'], axis=1), drop_first=True)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, ConfusionMatrixDisplay,classification_report
confusion_matrix(y_val, y_pred)
ConfusionMatrixDisplay.from_predictions(y_val, y_pred)
```
